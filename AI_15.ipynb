### 출력 크기 공식

# AlexNet

![Alexnet](imgs/AlexNet1.png)

구조(채널 수)

    Input → Conv (96) → Pool → Conv (256) → Pool → Conv (384) → Conv (384) → Conv (256) → Flatten(9216) → Dense(4096) → Dense(4096) → Output (10)

![Alexnet](imgs/AlexNet2.png)

딥러닝의 문제점 : 

1. 깊이가 깊어질수록 `chain-rule`에 의해 기울기가 0에 수렴해가는 `기울기 소실 문제`가 발생한다.
2. 과적합 문제 : 훈련 데이터에 지나치게 `과학습`되어 새로운 데이터에 극도로 민감하게 반응하는 현상

![Overfitting](imgs/Overfitting1.png)

해결책

1. ReLU 함수 사용

![ReLU](imgs/ReLU1.jpeg)

`기울기 소실 문제` 해결

2. DropOut

학습 중간 일부 결과값을 제거하여 과적합 방지

3. Data Augmentation

학습 데이터의 상하 반전, 좌우 반전 등 학습데이터 증강 기법
