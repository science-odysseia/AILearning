## 정규표현식

정규표현식 모듈

```python
import re
```

`.` : 아무 문자 1개(`\n`은 제외)

```python
r = re.compile("a.c")
print(r.search("kkk"))  # 아무런 결과도 출력되지 않는다.
print(r.search("abc"))
print(r.search("abdc")) # 아무런 결과도 출력되지 않는다.
print(r.search("ac"))   # 아무런 결과도 출력되지 않는다.
```

`?` : `?` 앞 문자 1개가 있을수도 없을수도 있음

```python
r = re.compile("ab?c")
print(r.search("ac"))
print(r.search("abc"))
print(r.search("abbc")) # 아무런 결과도 출력되지 않는다.
```

`*` : `*` 앞 문자가 0번이상 반복

```python
r = re.compile("ab*c")
print(r.search("a")) # 아무런 결과도 출력되지 않는다.
print(r.search("ac"))
print(r.search("abc"))
print(r.search("abbc"))
print(r.search("abbbc"))
```
